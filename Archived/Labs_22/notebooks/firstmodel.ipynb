```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
from google.colab import files
```

```python
train = pd.read_csv('sign_mnist_train.csv')
```

```python
train.shape
```

```python
test = pd.read_csv('sign_mnist_test.csv')
```

```python
train.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
features = train.columns
labels = np.array(train['label'].to_list())
features = features.drop('label')
features = train[features]
features = features/255
```

```python
train_features, val_features, train_labels, val_labels = train_test_split(features, labels, test_size=0.2)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential()
model.add(Dense(350, input_dim=784, activation='relu'))
model.add(Dense(26))

model.compile(optimizer='adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])
```

```python
callbacks = [EarlyStopping(monitor='val_acc', patience=10)]
```

```python
model.fit(train_features, train_labels, validation_data=(val_features, val_labels), callbacks=callbacks, epochs=100)
```

```python
test_features = test.columns
test_labels = np.array(test['label'].to_list())
test_features = test_features.drop('label')
test_features = test[test_features]
test_features = test_features/255
```

```python
test_loss, test_acc = model.evaluate(test_features,  test_labels, verbose=2)

print('\nTest accuracy:', test_acc)
```

```python
model.save('initial_model')
```
